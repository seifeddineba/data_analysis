Top 5 platform seals

```python
import matplotlib.pyplot as plt

data = pd.read_csv("vgsales.csv")

total_sales_by_platform = data.groupby('Platform')['Global_Sales'].sum()

top_five_platform_sales = total_sales_by_platform.sort_values(ascending=False).head(5)
# Plotting the bar chart
plt.bar(top_five_platform_sales.index, top_five_platform_sales.values)
plt.xlabel('Platform')
plt.ylabel('Total Sales')
plt.title('Total Sales by Platform')
plt.show()
```

Create a plot visualizing the average sales for games in the most popular three genres. Differentiate between NA, EU, and global sales.

```python
# Group the data by genre and calculate the average sales for NA, EU, and global markets
genre_sales = data.groupby('Genre')[['NA_Sales', 'EU_Sales', 'Global_Sales']].mean()

# Sort the genres by global sales in descending order and select the top 3
top_3_genres = genre_sales.sort_values(by='Global_Sales', ascending=False).head(3)

# Transpose the DataFrame to have genres as columns and sales types as rows
top_3_genres = top_3_genres.transpose()

# Plotting the bar chart
top_3_genres.plot(kind='bar')
plt.xlabel('Sales Type')
plt.ylabel('Average Sales')
plt.title('Average Sales for Games in Top 3 Genres')
plt.legend(title='Genre')
plt.show()
```

sales over the yeras

```python
# Group the data by year and calculate the sum of global sales for each year
sales_by_year = data.groupby('Year')['Global_Sales'].sum()

# Plotting the line chart
plt.plot(sales_by_year.index, sales_by_year.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Total Sales (in millions)')
plt.title('Total Sales by Year')
plt.grid(True)
plt.show()
```
